```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
modi_celeb = cv2.imread("/content/img.jpg")
```

```python
resized_modi = cv2.resize( modi_celeb,(620,500))
```

```python
cv2_imshow(resized_modi)
```

```python
#zooming into the flower
cropped_modi = resized_modi[100:450, 200:400]
cv2_imshow(cropped_modi)
```

```python
(a,b,c)=cropped_modi.shape[:3]
(x,y,z) = (a//2,b//2,c//2)
M=cv2.getRotationMatrix2D((x,y),55,1)
#translation of image 
Modi_rotate = cv2.warpAffine(cropped_modi,M,(a,b))
cv2_imshow(Modi_rotate)
```

```python
modi_red = Modi_rotate.copy()
modi_red[:,:,0]=0
modi_red[:,:,1]=0
```

```python
cv2_imshow(modi_red)
```

```python
hsv_frame = cv2.cvtColor(modi_celeb , cv2.COLOR_BGR2HSV)
```

```python
low_red = np.array([161, 155, 84])
high_red = np.array([179, 255, 255])
red_mask = cv2.inRange(hsv_frame, low_red, high_red)
red = cv2.bitwise_and(modi_celeb,modi_celeb, mask=red_mask)

# Blue color
low_blue = np.array([94, 80, 2])
high_blue = np.array([126, 255, 255])
blue_mask = cv2.inRange(hsv_frame, low_blue, high_blue)
blue = cv2.bitwise_and(modi_celeb, modi_celeb, mask=blue_mask)

# Green color
low_green = np.array([25, 52, 72])
high_green = np.array([102, 255, 255])
green_mask = cv2.inRange(hsv_frame, low_green, high_green)
green = cv2.bitwise_and(modi_celeb, modi_celeb, mask=green_mask)
```

```python


cv2_imshow(cv2.resize( red,(620,500)))

```
